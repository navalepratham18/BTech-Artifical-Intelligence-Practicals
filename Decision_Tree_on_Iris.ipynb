```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
iris = load_iris()    # Load the iris dataset

# convert too dataframe for easier handling
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

print("First five rows of the dataset:")
print(df.head())
```

```python
X = df.iloc[:, :-1]  # (select all rows and all columns except the last one)
Y = df['species']  # target colomn
```

```python
# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# create and train the Decision Tree Classifier
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Model Evaluation :")
print("-"*10)
print("Accuracy : ", accuracy_score(Y_test, y_pred))
print("Confusion Matric : \n", classification_report(Y_test, y_pred, target_names=iris.target_names))
```

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title("Decision Tree for Iris Dataset", fontsize=14)
plt.show()
```

```python
sample = [[5.1, 3.5, 1.4, 0.2]]
pred = model.predict(sample)
print(f"Predicted class for sample {sample} : {iris.target_names[pred][0]}")
```
